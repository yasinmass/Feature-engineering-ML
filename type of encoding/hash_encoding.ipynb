## Hash encoding 

Hash Encoding converts categories into numbers using a hash function.

```python
import pandas as pd 
import category_encoders as ce
```

```python
crop_data=pd.read_csv("dataset/Crop_recommendation.csv")
crop_data
```

```python
data = pd.DataFrame({
    'City': ['Chennai', 'Mumbai', 'Delhi', 'Chennai']
})
data
temp=data.copy()
```

```python
print(temp.shape)
temp.City.nunique()
```

```python
# Create Hash Encoder
encoder = ce.HashingEncoder(cols=['City'], n_components=3)
```

```python
# Apply encoding
encoded_data = encoder.fit_transform(data)
encoded_data.columns
```

```python
print(data.shape)
print(encoded_data.shape)
```

# When to Use

- When categorical feature has many unique values (high cardinality)
Example: User IDs, Product IDs, Zip Codes

- When One-Hot creates too many columns

- When memory efficiency is important

- Large-scale ML system

# Pros

- Fixed number of output columns

- Memory efficient

- Works well for large datasets

- Good for high-cardinality features

# Cons

- Hash collisions may occur

- Not interpretable

- Cannot reverse back to original category

- May slightly reduce model accuracy

*It is mainly used in production / Industry level with large dataset*

*it is used when our problem is dimension*
